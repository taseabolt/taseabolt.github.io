```python
#CS620 - Introduction to Data Science and Analytics
#Homework Exercise 2 (HW2): Spring 2026
#Tara Seabolt UIN:

# Import libraries
import pandas as pd
import numpy as np

#Q1 - Load diamonds.csv data file
url = "https://www.cs.odu.edu/~sampath/courses/data/diamonds.csv"
diamonds = pd.read_csv(url)
diamonds.head(5)

#Q2 - List the data types of df and list last 10 rows / items
diamonds.dtypes
diamonds.tail(10)

#Q3 - Convert the 'price' column to NumPy array 'diamond_price'
diamonds_price = diamonds['price'].to_numpy()

#Q3.1 - Middle diamond price / value in the 'diamond_price' array
mid_index = len(diamonds_price) // 2
middle_price = diamonds_price[mid_index]
print(f"Middle price : ${middle_price:,}")

#Q3.2 - 'diamond_price' array in ascending order
sorted_prices = np.sort(diamonds_price)
print("Sorted prices in ascending order:", sorted_prices)

#Q.3 - Range of the price lisetd in the 'diamond_price' array
price_range = int(sorted_prices[-1]) - int(sorted_prices[0])
print(f"Highest price: ${sorted_prices[-1]:,}")
print(f"Lowest price: ${sorted_prices[0]:,}")
print(f"Range: ${price_range:,}")

#Q4 - "Premium" type diamonds' depth, price, and clarity in that order
premium = diamonds[diamonds['cut'] == 'Premium'][['depth', 'price', 'clarity']]
print(f"Total # of Premium diamonds: {len(premium)}")
print(premium)

#Q5 - PriceVoltality value: 12% of the price
diamonds['priceVolatility'] = diamonds['price'].apply(lambda p: round(p * 0.12, 2))
print(list(diamonds.columns))

#Q5.1 - Display rows 4, 27, 1221 w/selected columns
diamonds.loc[[4, 27, 1221], ['carat', 'color', 'priceVolatility']]

#Q6 - Summary statistics for clairy 'VS1'
vs1 = diamonds[diamonds['clarity'] == 'VS1']
vs1.describe()
```
